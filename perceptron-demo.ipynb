```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/placement.csv')
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(df,x='cgpa',y='resume_score',hue=df['placed'])
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```
